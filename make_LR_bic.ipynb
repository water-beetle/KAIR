```python
import os
import glob
from PIL import Image
```

```python
files = glob.glob('C:/Users/kdh99/OneDrive/Desktop/drive-download-20221115T080830Z-001/manga109/manga109/*')
```

```python
for f in files:
    img = Image.open(f)
    img_high = img.resize((int(img.width / 2) * 2, int(img.height / 2) * 2))
    img_row = img.resize((int(img.width / 2), int(img.height / 2)))

    img_high.save("C:/Users/kdh99/OneDrive/Desktop/Manga109/HR/" + f.split('\\')[-1])
    img_row.save("C:/Users/kdh99/OneDrive/Desktop/Manga109/LR/" + f.split('\\')[-1])

    img.close()
```
