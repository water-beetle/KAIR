```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = ['base', '(s0,t1)+(s1,t3)', '(s0,t0)+(s1,t3)', '(s0,t0)', '(s1,t3)',
'(s0,t2)+(s1,t3)']
y = [31.24, 30.41, 30.59, 31.30, 30.79, 30.63] # set14
y1 = [36.21, 34.29, 34.67, 35.96, 35.28, 34.80] # Manga109
y2 = [29.49, 27.62, 27.83, 29.49, 28.16, 27.91] # Urban
```

```python
sns.set_theme(style="whitegrid")

ax = sns.barplot(x=x, y=y2)
plt.ylim(26, 30)
plt.xlabel('Loss + MSE(student, teacher)')
plt.ylabel('Average PSNR')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 20)

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 0.1, height, ha = 'center', size = 9)
```

```python
x = ['base', 'output_add']
y = [31.24, 31.28]
y1 = [36.21, 36.18]
y2 = [29.49, 29.56]

import pandas as pd

df = pd.DataFrame({'x' : ['base', 'base', 'base','output_add', 'output_add', 'output_add'],
'y' : [31.24, 36.21, 29.49, 31.28, 36.18, 29.56],
'dataset' : ['Set14', 'Manga109', 'Urban', 'Set14', 'Manga109', 'Urban']})
```

```python
g = sns.catplot(x='dataset', y='y', hue='x', data=df, kind='bar')
ax = g.facet_axis(0, 0)
plt.ylim(28, 37)
plt.ylabel('Average PSNR')
plt.xlabel('Dataset')
for c in ax.containers:
    labels = [f'{(v.get_height()):.2f}' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge')
```
